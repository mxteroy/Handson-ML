# Gradient Boosting

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score # only supports classification tasks

boston = load_boston()
X, y = boston.data, boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

dtree_clf1 = DecisionTreeRegressor(max_depth=2)
dtree_clf1.fit(X_train, y_train)
y_pred = dtree_clf1.predict(X_train)
# accuracy_score(y_train, y_pred)
y_train1 = y_train - y_pred
dtree_clf1.score(X_test, y_test)
```

```python
dtree_clf2 = DecisionTreeRegressor(max_depth=2)
dtree_clf2.fit(X_train, y_train1)
y_train2 = y_train1 - dtree_clf2.predict(X_train)
```

```python
dtree_clf3 = DecisionTreeRegressor(max_depth=2)
dtree_clf3.fit(X_train, y_train2)
```

```python
y_pred = sum(tree.predict(X_test) for tree in (dtree_clf1, dtree_clf2, dtree_clf3)) # the predictions for the input dataset
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.sort(X_test[:, 0]),  y_test)
plt.plot(np.sort(X_test[:, 0]), y_pred)
plt.show()
```
