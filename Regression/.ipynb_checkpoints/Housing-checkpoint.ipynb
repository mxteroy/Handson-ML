```python
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
        
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path) #url, filename
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
    
fetch_housing_data()
```

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

```python
import numpy as np

def split_train_test(data, test_ratio):
    N = len(data)
    np.random.seed(42)
    shuffled_indices = np.random.permutation(N)
    test_set_size = int(test_ratio * N)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size]
    return data.iloc[test_indices], data.iloc[train_indices]

test_set, train_set = split_train_test(housing, .2)
```

```python
import hashlib

def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    print(type(data))
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    print(in_test_set, ~in_test_set)
    return data.loc[~in_test_set], data.loc[in_test_set] #loc() return values in indexes where boolean array is True

housing_with_id = housing.reset_index() #adds index column. only works if new data is appended. no existing data gets deleted/replaced
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"] #better give ids
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

```python
'''
Need to create custom transformers class with three methods (fit(), transofrm(), and fit_transform()) so that it can work with pipelines and other scikitlearn
functionalities
- Inherting TransformerMixin class will add the fit_transform() method automatically
- Inheriting BaseEstimator will give bonus get_params() and set_params() methods (useful for auto hyperparameter tuning)
'''

from sklearn.base import BaseEstimator, TransformerMixin

# indexes in housing
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): 
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y = None):
        return self #nothing else to do
    
    def transform(self, X, y = None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, population_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                       bedrooms_per_room]
        
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room = False) # add_bedrooms_per_room is hyperparameter
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
'''
Feature scaling
ML algorithms don't perform well when numerical attributes have very differnt scales

Two ways to feature scale: min-max scaling (normalization; change ranges of values to 0 to 1) and standardization
'''

```

```python
'''
Transformation Pipelines

Do all of the transformation steps above in much less code by using pipelines
'''

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


# last estimator must be transfomers (has fit_transform() method)
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])
```

```python
'''
DataFrameSelector to extract the numerical columsn in to a NP array
'''

from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X[self.attribute_names].values
    
    
    
```

```python
num_attribs = list(housing_num) # making a DF into a list only returns the features not the feature-values

cat_attribs = ["ocean_proximity"]
# add selector to pick out numerical attributes from DataFrame (get rid of the non-numerical)
num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

# pipeline to get the categories
cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('cat_encoder', OneHotEncoder())
])
```

```python
'''
Combine two pipelines together
'''

from sklearn.pipeline import FeatureUnion

full_pipeline = FeatureUnion(transformer_list=[
    ("num_pipeline", num_pipeline),
    ("cat_pipeline", cat_pipeline)
])

housing_prepared = full_pipeline.fit_transform(housing)
```

```python
from sklearn.metrics import mean_squared_error

def predict_and_get_mse(model, data, labels):
    housing_predictions = model.predict(data)
    lin_mse = mean_squared_error(labels, housing_predictions)
    lin_rmse = np.sqrt(lin_mse)
    return lin_rmse

print(predict_and_get_mse(lin_reg, housing_prepared, housing_labels)) #huge prediction of median housing value. Off by $68898!
```
